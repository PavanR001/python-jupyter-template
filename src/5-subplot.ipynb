### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

try:
    # Data for plots
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    categories = ['A', 'B', 'C', 'D']
    values = [10, 15, 7, 12]
    scatter_x = np.random.rand(50)
    scatter_y = np.random.rand(50)
    pie_values = [30, 20, 25, 25]
    pie_labels = ['Red', 'Blue', 'Green', 'Yellow']

    # Validate data before plotting
    if any(np.isnan(y)) or len(x) == 0 or len(y) == 0:
        raise ValueError("Invalid data: NaN values detected or empty arrays.")

    # Create figure and subplots
    fig, axs = plt.subplots(2, 2, figsize=(10, 8))

    # Line chart
    try:
        axs[0, 0].plot(x, y, label='sin(x)', color='blue')
        axs[0, 0].set_title('Line Chart')
        axs[0, 0].legend()
    except ValueError as e:
        print(f"Error in Line Chart: {e}")

    # Bar chart
    try:
        axs[0, 1].bar(categories, values, color='orange')
        axs[0, 1].set_title('Bar Chart')
    except ValueError as e:
        print(f"Error in Bar Chart: {e}")

    # Scatter plot
    try:
        axs[1, 0].scatter(scatter_x, scatter_y, color='green')
        axs[1, 0].set_title('Scatter Plot')
    except ValueError as e:
        print(f"Error in Scatter Plot: {e}")

    # Pie chart
    try:
        axs[1, 1].pie(pie_values, labels=pie_labels, autopct='%1.1f%%', colors=['red', 'blue', 'green', 'yellow'])
        axs[1, 1].set_title('Pie Chart')
    except ValueError as e:
        print(f"Error in Pie Chart: {e}")

    # Adjust layout and display plot
    plt.tight_layout()
    plt.show()

except ValueError as ve:
    print(f"Value Error: {ve}")
except ImportError:
    print("Error: Matplotlib or NumPy is not installed.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
